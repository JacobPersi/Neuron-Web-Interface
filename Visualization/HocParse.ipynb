# Hacky Neuron 'Hoc' file parser:
A simple script to extract the morphology data for a specific hoc file. 
Note: This parsing is based on the way this file is formatted and so will not work for all hoc files.

```python
hoc_file = open("C:/Users/Jacob/Neural-Modelling/Morphology/1148.hoc", "r")
lines = hoc_file.readlines()

index = 0
shape_open = False
shapes = []
section_open = False

for line in lines:
    # Start of shape:
    if "proc shape3d_" in line:
        shape_open = True
        shapes.append({})
        pass
    # End of shape:
    if line == "}\n":
        shape_open = False
        pass

    # Inside the shape:
    if shape_open == True:
        # Start of section:
        if "{" in line and not "proc" in line:
            section_open = True
            section_name = line[:line.index("{")].strip()
            shapes[len(shapes)-1][section_name] = []
            # We don't pass here because some shapes have the name and points on the same line.

        # Inside of the section:
        if section_open == True:
            if "pt3dadd" in line:
                coordinates = line[line.index("pt3dadd(") + len("pt3dadd("):-2].strip()
                coordinates = [float(x) for x in coordinates.split(", ")]
                diameter = coordinates.pop()
                coordinates = tuple(coordinates)
                shapes[len(shapes)-1][section_name].append(coordinates)
                pass

        # End of section:
        if line == "  }\n":
            section_open = False
            pass

    index += 1
```

```python
for shape in shapes:
    for section in shape:
        vertices = shape[section]
        edges = [(i, i + 1) for i in range(0,len(shape[section])-1,1)]

        print("Name:", section)
        print("Vertices:", vertices)
        print("Edges:", edges)
        print()

    break
```
